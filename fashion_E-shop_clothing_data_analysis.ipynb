** maximuim volume of sales (per period and per month)**

```python
def sum_sale(dataframe):
  orders_per_day = {}
  orders_per_month = {}
  for a in dataframe['year'].unique():
    for m in dataframe['month'].unique():
      sm = 0
      for j in dataframe['day'].unique():
        sj = len(dataframe['order'].loc[(dataframe['month'] == m) & (dataframe['day'] == j) & (dataframe['year']==a)])
        if sj != 0:
          orders_per_day[str(j)+"-"+str(m)+"-"+str(a)] = sj
        sm += sj  
      orders_per_month[m] = sm
  return orders_per_day, orders_per_month
```

```python
def max_sale(d):
  max = {}
  m = 0
  date = ""
  for k, v in d.items():
    if v>=m:
      m = v
      date = k
  max[date] = m
  return max
```

** More sales in first/second/thrid 10days of the month **

```python
def max_periond_sale(dm,dd):
  months = dm.keys()
  list_max = []
  list_pp = []
  for m in months:
    period_per_month = {}
    pp = {}
    sb = sm = se = 0
    for k, v in dd.items():
      l = k.split("-")
      if int(l[1]) == m:
        if int(l[0]) in range(1,11):      
          sb +=v   
        if int(l[0]) in range(11,21):
          sm +=v
        if int(l[0]) in range(21,31):
          se +=v
    maximum = sb
    maximum_p = "begin_month"
    if sm>maximum:
      maximum = sm
      maximum_p = "middle_month"
    elif se>maximum:
      maximum = se
      maximum_p = "end_month"
    period_per_month["month"] = m
    period_per_month["max period sale"] = maximum_p
    period_per_month["number of transactions"] = maximum
    pp["month"] = m
    pp["begin_month"] =sb
    pp["middle_month"] =sm
    pp["end_month"] =se
    list_pp.append(pp)
    list_max.append(period_per_month)
    nb_begin = nb_middle = nb_end = 0
    for item in list_max:
      if item["max period sale"] == "begin_month":
        nb_begin +=1
      elif item["max period sale"] == "middle_month":
        nb_middle +=1
      elif item["max period sale"] == "end_month":
        nb_end +=1
    d1 = {"begin_month":nb_begin,"middle_month":nb_middle,"end_month":nb_end}
    mmx = d1["begin_month"]
    period = "begin_month"
    for k,v in d1.items():
      if v>mmx:
        mmx = v
        period= k
  return list_max,list_pp,mmx,period
```

```python
list_max,list_pp,mmx,period = max_periond_sale(orders_per_month,orders_per_day)
print(list_pp)
print(list_max)
print("The most sales is in the ",period)
```

```python
def search_max(dx):
  m = 0
  for k,v in dx.items():
    if v>m:
      m = v
      mc = k
  return m,mc
def combinaison_result(dataframe,c1,c2):
  res = {}
  for c in dataframe[c2].unique():
    res[c] = dataframe[c1].loc[dataframe[c2] == c]
  return res
def maximum_combinaison(c1,c2,ds,c):
  d1 = ds
  max = {c2:c,c1:[]}
  for i in range(3):
    m,mc = search_max(d1)
    max[c1].append(mc)
    d1.pop(mc)
  return max
def show_combinaison_result(dataframe,c1,c2):
  d = combinaison_result(dataframe,c1,c2)
  
  for k in d.keys():
    dv = dict(d[k].value_counts())
    print(maximum_combinaison(c1,c2,dv,k))
```

```python
def max_number_sale_customer(dataframe):
  achat_par_client = {}
  for i in dataframe['session ID'].unique():
    l = []
    l.append(nationalities[dataframe['country'].loc[dataframe['session ID'] == i].unique()[0]])
    for item in list(dataframe['page 2 (clothing model)'].loc[dataframe['session ID'] == i]):
      l.append(item)
    achat_par_client[i] = l
  max = 0
  for k,v in achat_par_client.items():
    if len(v) -1 > max:
      max = len(v) -1
  return max
```

```python
max_all = max_number_sale_customer(df)
max_tun = max_number_sale_customer(df_tun)
max_euro = max_number_sale_customer(df_european)
max_non_euro = max_number_sale_customer(df_non_european)
```

```python
def prepare_list(dataframe,max):
  list_columns = []
  for i in range(0, max):
    list_columns.append(i+1)
  list_achat = []
  for i in dataframe['session ID'].unique():
    l = []
    cp = 1
    for item in list(dataframe['page 2 (clothing model)'].loc[dataframe['session ID'] == i]):
      l.append(item)
      cp += 1
    for j in range(max - len(l)):
      l.append(None)
      cp +=1
    list_achat.append(l)
  ar = np.array(list_achat)
  df_result = pd.DataFrame(ar, columns = list_columns)
  return df_result
```

```python
def get_transaction(dataframe,max):
  transactions = []
  for i in range(0, len(dataframe)):
    transactions.append([str(dataframe.values[i,j]) for j in range(0, max) if str(dataframe.values[i,j])!="None"])
  return transactions
```

```python
transactions_all = get_transaction(df_res_all,max_all)
transactions_tun = get_transaction(df_res_tun,max_tun)
transactions_euro = get_transaction(df_res_euro,max_euro)
transactions_non_euro = get_transaction(df_res_non_euro,max_non_euro)
```

```python
def association_rules(transactions,min_supp,min_conf,min_lef):
  rules = apriori(transactions, min_support = min_supp, min_confidence = min_conf, min_lift = min_lef)
  results = list(rules)
  return results
```

```python
results_all = association_rules(transactions_all,min_supp=0.02,min_conf=0.45,min_lef=1)
results_tun = association_rules(transactions_tun,min_supp=0.02,min_conf=0.45,min_lef=1)
results_euro = association_rules(transactions_euro,min_supp=0.045,min_conf=0.48,min_lef=3)
results_non_euro = association_rules(transactions_non_euro,min_supp=0.013,min_conf=0.45,min_lef=1)
```

```python
def inspect(results):
    li          = [tuple(result[2][0][0])[0] for result in results]
    ri          = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(li, ri, supports, confidences, lifts))
```

```python
def get_resultsinDataFrame(results):
  return pd.DataFrame(inspect(results), columns = ['Left Item', 'Right Item', 'Support', 'Confidence', 'Lift'])
```

```python
resultsinDataFrame_all = get_resultsinDataFrame(results_all)
resultsinDataFrame_tun = get_resultsinDataFrame(results_tun)
resultsinDataFrame_euro = get_resultsinDataFrame(results_euro)
resultsinDataFrame_non_euro = get_resultsinDataFrame(results_non_euro)
```
